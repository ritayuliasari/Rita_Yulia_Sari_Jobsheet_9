```python
import math

def hitung_luas_lingkaran(jari_jari):
  """
  Menghitung luas lingkaran.
  """
  if jari_jari < 0:
    raise ValueError("Jari-jari tidak boleh negatif")

  luas = math.pi * (jari_jari ** 2)
  return luas


input_jari_jari = input("Masukkan nilai jari-jari lingkaran: ")

try:
  jari_jari = float(input_jari_jari)

  luas_hasil = hitung_luas_lingkaran(jari_jari)

  print("1.Eksperimen Dengan Modul Math")
  print(f"\nHasil Perhitungan:")
  print(f"Nilai Jari-jari (r): {jari_jari}")
  print(f"Nilai Pi (π): {math.pi}")
  print(f"Luas Lingkaran (L = π * r²): {luas_hasil}")

except ValueError as e:
  if "could not convert string" in str(e):
    print("\nError: Masukkan harus berupa angka yang valid.")
  else:
    print(f"\nError: {e}")
```

```python
%%writefile konversi.py
def celcius_ke_fahrenheit(c):
  return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
  return (f - 32) * 5/9
```

```python
import konversi

print("4.Modul Buatan Sendiri")

suhu_celcius = 100
suhu_fahrenheit = 212

hasil_fahrenheit = konversi.celcius_ke_fahrenheit(suhu_celcius)
hasil_celcius = konversi.fahrenheit_ke_celcius(suhu_fahrenheit)

print(f"{suhu_celcius} Celcius adalah {hasil_fahrenheit} fahrenheit")
print(f"{suhu_fahrenheit} Fahrenheit adalah {hasil_celcius} celcius")
```

```python
%%writefile laporan.py

def cetak_laporan(skor, tottal_soal):
  print("\n--- KUIS SELESAI ---")
  print(f"Skor Anda: {skor}/{totatal_soal}")

  persentase = (skor / tottal_soal) * 100
  print(f"Nilai Akhir: {persentase:.if}")

  if persentase == 100:
    print("Selamat! Anda mendapat nilai sempurna.")
  elif persentase >= 70:
    print("Anda lulus ujian dengan baik.")
  else:
    print("Mohon maaf, Anda belum lulus ujian.")
```
